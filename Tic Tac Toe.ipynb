```python
# Makes board look better

def boardbox(board):
    print(str(board[0][0]) + '|' + str(board[0][1]) + '|' + str(board[0][2]))
    print('-+-+-')
    print(str(board[1][0]) + '|' + str(board[1][1]) + '|' + str(board[1][2]))
    print('-+-+-')
    print(str(board[2][0]) + '|' + str(board[2][1]) + '|' + str(board[2][2]))
```

```python
# player will enter number, appropriate position should be updated

def update_board (num, xo, board):
    if num == 1:
        board[0][0] = xo
    elif num == 2:
        board[0][1] = xo
    elif num == 3:
        board[0][2] = xo
    elif num == 4:
        board[1][0] = xo
    elif num == 5:
        board[1][1] = xo
    elif num == 6:
        board[1][2] = xo
    elif num == 7:
        board[2][0] = xo
    elif num == 8:
        board[2][1] = xo
    elif num == 9:
        board[2][2] = xo
# add if position not an integer, then you can fill it.
```

```python
#check if position is empty is not.

def position_available (num, board):
    if num == 1:
        position = board[0][0] 
    elif num == 2:
        position = board[0][1]
    elif num == 3:
        position = board[0][2]
    elif num == 4:
        position = board[1][0]
    elif num == 5:
        position = board[1][1]
    elif num == 6:
        position = board[1][2]
    elif num == 7:
        position = board[2][0]
    elif num == 8:
        position = board[2][1]
    elif num == 9:
        position = board[2][2]
    if position == ' ':
        return True
    else:
        return False
```

```python
# updating board in each turn only if position available to avoid overwriting previous choice
# won't check player 2 vs player 1. xo identifies these

def turn(num, xo, board):
    if position_available(num,board) == True:
        update_board(num, xo, board)
    else:
        print('Position not available. Try again.')
```
